```python
import requests
import json
import pandas as pd
```

```python
url = "https://imdb-top-100-movies.p.rapidapi.com/"
```

```python
headers = {
    "x-rapidapi-key": "",
    "x-rapidapi-host": ""
}
```

```python
response = requests.get(url, headers=headers)
response_json = response.json()
print(response_json)
```

```python
top_movies_df = pd.DataFrame(response_json)
top_movies_df.head()
```

```python
top_movies_df.info()
```

```python
top_movies_df.isnull().sum()
```

```python
## Transformations

# Extracting required columns 
top_movies_df = top_movies_df[['rank', 'title', 'description', 'image', 'genre', 'rating', 'id', 'year', 'imdbid', 'imdb_link']]

# Renaming columns for clarity
top_movies_df.rename(columns={
    'image': 'image_url'
}, inplace=True)
```

```python
top_movies_df.head()
```

```python
# Category bins for movie year

top_movies_df['year'].unique()
```

```python
bins = [1900, 1980, 1990, 2000, 2010, 2020]
labels = ['Before 1980', '1980-1990', '1990-1999', '2000-2009', '2010-2020']
top_movies_df['year_category'] = pd.cut(top_movies_df['year'], bins=bins, labels=labels)
```

```python
top_movies_df.head()
```

```python
top_movies_df['genre'].head()
```

```python
top_movies_df['genre_flattened'] = top_movies_df['genre'].apply(lambda x: ', '.join(x))
top_movies_df.head()
```

```python
top_movies_df['genre_count'] = top_movies_df['genre'].apply(len)
top_movies_df.head()
```
